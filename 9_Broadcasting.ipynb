## Broadcasting 

whenever we apply operations on two arrays such as addition & substraction, we always consider that dimensions of arrays are same. but what if the dimensions are not same instead they are different, that's where broadcating comes into the picture :-

here the explanation is quite difficult so I prefer you to watch this video from campusx on broadcasting :-

```python
import numpy as np

arr1 = np.arange(12).reshape(3, 4)
arr2 = np.array([10, 20, 30, 40])

print(arr1 + arr2,'\n')

# lets see the other cases of broadcasting
arr3 = np.array([[1], [2], [3]])
print(arr1 + arr3)

# another example
arr4 = np.array([[100], [200], [300], [400]])
print('\n', arr2 + arr4)

# now this one is specially interesting
arr5 = np.array([1, 2, 3])

# print('\n', arr1 + arr5)  
# this will raise an error due to incompatible shapes

# but if we reshape arr5 to (3,1) it will work
arr5_reshaped = arr5.reshape(3, 1)
print('\n', arr1 + arr5_reshaped)

# now if arr6 is (4,1) but arr7 is (1,4), then it will work too
arr6 = np.array([[1], [2], [3], [4]])
arr7 = np.array([[10, 20, 30, 40]])
print('\n', arr6 + arr7)
```
