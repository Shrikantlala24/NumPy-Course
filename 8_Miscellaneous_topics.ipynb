## 1. Fancy Indexing
when there is a matrix (2D array), in order to slect custom rows and columns we use fance indexing.

```python
import numpy as np

arr1 = np.arange(24).reshape(6,4)
print(arr1,'\n')
print(arr1[[0,3,5]])
```

## 2. Boolean masking
when we put conditions in the array it returns a boolean array. now if put the condition statements as indexing in the array, it will give you the elements of array which satisfy the condition.

now for boolean masking, we equate the statement with the desired value (used to eqaute with "0")

```python
arr2 = np.random.randint(25,50,size = (5,5))
print(arr2,'\n')

print(arr2>35,'\n')
print(arr2[arr2>35],'\n')

arr2[arr2>35] = 0
print(arr2)
```
