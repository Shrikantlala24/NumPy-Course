# Array object attributes :-
### 1. .shape

```python
import numpy as np

arr = np.array([1,2,3,4,5,6])
print(arr)

arr.shape
```

### 2. .ndim

```python
arr2 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(arr2)

arr2.ndim
```

### 3. .size

```python
arr3 = np.random.randint(3,4,size=(3,3))
print(arr3)

print(arr3.size)
# check for the orevious one
arr2.size 
```

### 4. .itemsize

```python
arr4 = np.random.rand(3,2)
print(arr4)

print(arr4.itemsize)

# let's convert the array in integer and check the itemsize
print(arr4.astype('int').itemsize)
```

### 5. .dtype

```python
print(arr3.dtype)

print(arr4.dtype)
```
