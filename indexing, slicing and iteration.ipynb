# Indexing in arrays and matrices
## 1. Indexing in arrays

```python
import numpy as np

arr1 = np.arange(0,16).reshape(4,4)
print(arr1,"\n")

print(arr1[3])
print(arr1)
```

```python
arr3 = np.arange(16).reshape(4,4)
print("\n",arr3)

arr4 = np.array_split(arr3,3)
arr4

arr3[[1,3]]

arr5 = np.random.randint(1,10, size= (2,2))
print(arr5)
arr5[arr5 >5]=0
arr5
```

```python
x = np.linspace(-10,10,30)
y = np.sin(x)

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x,y)
plt.show()
```
