## Creating arrays with built-in methods :-

```python
import numpy as np
```

### 1. arange(start, stop, step)

```python
arr = np.arange(0,5,1)
print(arr)
```

### 2. zeros(vals) or zeros((rows,cols))

```python
arr2 = np.zeros((5,5))
arr2
```

### 3. Ones() {same params as above}

```python

arr3_1 = np.ones(4)
print(arr3_1,"\n")


arr3_2 = np.ones((3,4))
print(arr3_2)
```

### 4. identity(dim) or eye(dim)

```python
arr4 = np.identity(5)
arr4
```

### 5. linspace(start, stop, divisions)

```python
arr5 = np.linspace(10,20,11)
arr5
```

### 6. rand() {same v or r,c}

```python
arr6 = np.random.rand(2,3)
arr6
```

### 7. randn() {same}

```python
arr7 = np.random.randn(5)
arr7
```

### 8. randint(start, stop, size = (r,c)) 

```python
arr8 = np.random.randint(1,4,size=(3,3))
arr8
```

### 9. a = b.copy()

```python
arr9 = np.copy(arr8)
arr9
```
