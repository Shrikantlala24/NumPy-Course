# Numpy Arrays VS Python lists
there are three advantages we have in using numpy arrays instead of python's built-in Lists :-

    1. Arrays takes less memory then lists.
let's prove it :-


```python
import numpy as np
import sys

# here you'll get the size in bytes of a list of integers
l1 = [x for x in range(100)]
print(sys.getsizeof(87) * len(l1))

## here you'll get the size in bytes of a numpy array
arr1 = np.arange(100)
print(arr1.itemsize * arr1.size)
```

---
    2. arrays are much faster in reading, writing and operating as compared to Lists.
    3. It's too easy to apply operations on arrays then lists
let's give it a try :-


```python
import time

a = 10000000
b = 20000000

l2 = [x for x in range(a)]
l3 = [x for x in range(a,b)]

start_time = time.time()

sum_list = [x + y for x, y in zip(l2, l3)]

# here you'll get the ime taken to perform addition on lists
print(time.time() - start_time)
```

```python
arr2 = np.arange(a)
arr3 = np.arange(a,b)

start_time = time.time()

sum_array = arr2 + arr3

# here you'll get the time taken to perform addition on numpy arrays
print(time.time() - start_time)
```
